```python
# Ensure Memory column is string and clean base units
df['Memory'] = df['Memory'].astype(str).replace(r'\.0', '', regex=True)
df["Memory"] = df["Memory"].str.replace('GB', '', regex=False)
df["Memory"] = df["Memory"].str.replace('TB', '000', regex=False)

# Split into first + second part
new = df["Memory"].str.split("+", n=1, expand=True)
df["first"] = new[0].str.strip()
df["second"] = new[1].fillna("0")  # fill missing with "0"

# Detect storage type for first layer
df["Layer1HDD"] = df["first"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer1SSD"] = df["first"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer1Hybrid"] = df["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

# Detect storage type for second layer
df["Layer2HDD"] = df["second"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["second"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

# Keep only digits for numeric conversion
df["first"] = df["first"].str.replace(r'\D', '', regex=True).replace('', '0').astype(int)
df["second"] = df["second"].str.replace(r'\D', '', regex=True).replace('', '0').astype(int)

# Create final numeric columns
df["HDD"] = (df["first"] * df["Layer1HDD"] + df["second"] * df["Layer2HDD"])
df["SSD"] = (df["first"] * df["Layer1SSD"] + df["second"] * df["Layer2SSD"])
df["Hybrid"] = (df["first"] * df["Layer1Hybrid"] + df["second"] * df["Layer2Hybrid"])
df["Flash_Storage"] = (df["first"] * df["Layer1Flash_Storage"] + df["second"] * df["Layer2Flash_Storage"])

# Drop helper columns
df.drop(columns=[
    'first', 'second',
    'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid', 'Layer1Flash_Storage',
    'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid', 'Layer2Flash_Storage'
], inplace=True)
```

# Random Forest

```python
# Step 1: Preprocessing (OneHotEncoder)
step1 = ColumnTransformer(transformers=[
   ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0,1,7,10,11])
], remainder='passthrough')

# Step 2: Model
step2 = RandomForestRegressor(random_state=3)

# Pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Hyperparameter grid
param_grid = {
    'step2__n_estimators': [100, 200, 300, 500],
    'step2__max_depth': [None, 10, 15, 20, 30],
    'step2__max_features': ['sqrt', 'log2', 0.5, 0.75, 1.0],
    'step2__min_samples_split': [2, 5, 10],
    'step2__min_samples_leaf': [1, 2, 4],
    'step2__bootstrap': [True, False]
}

# RandomizedSearchCV
search = RandomizedSearchCV(
    estimator=pipe,
    param_distributions=param_grid,
    n_iter=20,  # try 20 random combinations
    cv=3,
    scoring='r2',
    n_jobs=-1,
    verbose=2,
    random_state=42
)

# Fit
search.fit(X_train, y_train)

# Best model
best_model = search.best_estimator_

# Predictions
y_pred = best_model.predict(X_test)

# Evaluation
print("Best Parameters:", search.best_params_)
print("R2 Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
```

# Stacking

```python

# Step 1: Preprocessig
step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0,1,7,10,11])
], remainder='passthrough')

# Step 2: Base models
estimators = [
    ('rf', RandomForestRegressor(random_state=3)),
    ('gbdt', GradientBoostingRegressor(random_state=3)),
    ('xgb', XGBRegressor(random_state=3, verbosity=0))  # silence warnings
]

# Stacking Regressor
stack = StackingRegressor(estimators=estimators, final_estimator=Ridge(alpha=100))

# Pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', stack)
])


# Hyperparameter Grid
param_grid = {
    # Random Forest
    'step2__rf__n_estimators': [200, 350, 500],
    'step2__rf__max_depth': [10, 15, None],
    'step2__rf__max_features': [0.5, 0.75, 'sqrt'],
    'step2__rf__bootstrap': [True, False],

    # Gradient Boosting
    'step2__gbdt__n_estimators': [100, 200, 300],
    'step2__gbdt__learning_rate': [0.05, 0.1, 0.2],
    'step2__gbdt__max_depth': [3, 5, 7],
    'step2__gbdt__subsample': [0.6, 0.8, 1.0],

    # XGBoost
    'step2__xgb__n_estimators': [50, 100, 200],
    'step2__xgb__learning_rate': [0.05, 0.1, 0.3],
    'step2__xgb__max_depth': [3, 5, 7],
    'step2__xgb__subsample': [0.6, 0.8, 1.0],
    'step2__xgb__colsample_bytree': [0.6, 0.8, 1.0]
}

# RandomizedSearchCV
search = RandomizedSearchCV(
    estimator=pipe,
    param_distributions=param_grid,
    n_iter=30,            # increase for better search
    cv=3,                 # cross-validation
    scoring='r2',
    n_jobs=-1,
    verbose=2,
    random_state=42
)

# Fit
search.fit(X_train, y_train)

# Best Model
best_model = search.best_estimator_

y_pred = best_model.predict(X_test)

print("✅ Best Params:", search.best_params_)
print("✅ R2 Score:", r2_score(y_test, y_pred))
print("✅ MAE:", mean_absolute_error(y_test, y_pred))

```

# Voting Regressor

```python
# Step 1: Preprocessing
step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0,1,7,10,11])
], remainder='passthrough')

# Step 2: Base models
rf = RandomForestRegressor(random_state=3)
gbdt = GradientBoostingRegressor(random_state=3)

# Voting Regressor
voting = VotingRegressor([('rf', rf), ('gbdt', gbdt)])

# Pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', voting)
])

# Hyperparameter grid
param_grid = {
    # RandomForest params
    'step2__rf__n_estimators': [100, 200, 350],
    'step2__rf__max_depth': [10, 15, 20],
    'step2__rf__max_features': [0.5, 0.75, 'sqrt'],
    'step2__rf__min_samples_split': [2, 5],
    
    # GradientBoosting params
    'step2__gbdt__n_estimators': [100, 200, 300],
    'step2__gbdt__learning_rate': [0.05, 0.1, 0.2],
    'step2__gbdt__max_depth': [3, 5, 7],
    'step2__gbdt__subsample': [0.6, 0.8, 1.0]
}

# RandomizedSearchCV
search = RandomizedSearchCV(
    estimator=pipe,
    param_distributions=param_grid,
    n_iter=20,
    cv=3,
    scoring='r2',
    n_jobs=-1,
    verbose=2,
    random_state=42
)

# Fit
search.fit(X_train, y_train)

# Best model
best_model = search.best_estimator_

# Predictions
y_pred = best_model.predict(X_test)

# Evaluation
print("Best Parameters:", search.best_params_)
print("R2 Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
```
